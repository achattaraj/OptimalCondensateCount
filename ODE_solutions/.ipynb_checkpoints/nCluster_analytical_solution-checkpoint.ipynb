```python
import numpy as np
from numpy import array, exp, sqrt  
import matplotlib.pyplot as plt


font = {'family' : 'Arial',
        'size'   : 16}

plt.rc('font', **font)


n = np.arange(1,1000,1)

# Initial condition

G0 = 100
F0 = 0.01 
C = G0 + F0

# rate parameters 

k1=0.001
k2=1

# Arp2/3 density 
a = 1


title = f'C = {C}, k1 = {k1}, k2 = {k2}, a = {a}'

# per cluster F production in the n-cluster system

fn = lambda n, a, C, k1, k2: C - sqrt(k2/(k1*a))*n**(1/6)

Fn = fn(n, a=a, C=C, k1=k1, k2=k2)

F_global = (n*Fn)

plt.plot(np.log10(n), np.log10(Fn), 'mo', ms=4)
plt.xticks([0,1,2,3], labels=['10$^0$','10$^1$','10$^2$','10$^3$'])
plt.yticks([-1,0,1,2], labels=['10$^{-1}$','10$^0$','10$^1$','10$^2$'])
#plt.plot(n, Fn, 'bo', ms=4)
plt.xlabel('Cluster number (n)', fontweight='bold')
plt.ylabel('Local F (F$_n$)')
```

```python
Fn = fn(n, a=a, C=C, k1=k1, k2=k2)

F_global = (n*Fn)

idx = np.argmax(F_global)

plt.plot(np.log10(n), np.log10(F_global), 'go', ms=5)
plt.axvline(np.log10(n[idx]), ls='dashed', lw=2, c='k')
plt.xlabel('Cluster number (n)', fontweight='bold')
plt.ylabel('Global F (n * F$_n$)')
plt.xticks([0,1,2,3], labels=['10$^0$','10$^1$','10$^2$','10$^3$'])
plt.yticks([1,2,3,4], labels=['10$^1$','10$^2$','10$^3$','10$^4$'])
plt.tight_layout()
```
